```python
import pandas as pd
# to enable plotting graph in jupyter notebook
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
%matplotlib inline
```

```python
import os
```

```python
#selecting the directory in which dataset is present
os.listdir('../input/youtube-new')
```

```python
#loading dataset
invideos_df = pd.read_csv('../input/youtube-new/INvideos.csv')
```

```python
# exploring the dataset
invideos_df
```

```python
#checking the total no of rows and columns
invideos_df.shape
```

```python
# finding the values that are already repeated and removing it
invideos_df.drop_duplicates(subset='title', keep='first', inplace=True)
invideos_df
```

```python
#listing all the columns in the dataset
invideos_df.columns
```

```python
#checking the datatype 
invideos_df.info()
```

```python
#converting some of the int and boolean type to numeric format
invideos_df['views'] = pd.to_numeric(invideos_df.views, errors='coerce')
invideos_df['likes'] = pd.to_numeric(invideos_df.likes, errors='coerce')
invideos_df['dislikes'] = pd.to_numeric(invideos_df.dislikes, errors='coerce')
invideos_df['comment_count'] = pd.to_numeric(invideos_df.comment_count, errors='coerce')
invideos_df['comments_disabled'] = pd.to_numeric(invideos_df.comments_disabled, errors='coerce')
invideos_df['ratings_disabled'] = pd.to_numeric(invideos_df.ratings_disabled, errors='coerce')
invideos_df['video_error_or_removed'] = pd.to_numeric(invideos_df.video_error_or_removed, errors='coerce')
```

```python
#converting object type into datetime format
invideos_df['publish_time'] = pd.to_datetime(invideos_df.publish_time)
```

```python
#From publish time, create new columns of year,month,day,weekday and hour 
invideos_df['year'] = pd.DatetimeIndex(invideos_df.publish_time).year
invideos_df['month'] = pd.DatetimeIndex(invideos_df.publish_time).month
invideos_df['day'] = pd.DatetimeIndex(invideos_df.publish_time).day
invideos_df['weekday'] = pd.DatetimeIndex(invideos_df.publish_time).weekday
invideos_df['hour']=pd.DatetimeIndex(invideos_df.publish_time).hour
```

```python
#replacing november 2017 month as -1 and december 2017 as 0
invideos_df['month'] invideos_df['month'].replace({11:-1, 12:0})
#replacing true as 1 and false as 0
invideos_df['comments_disabled']=invideos_df['comments_disabled'].replace({True:1,False:0})
invideos_df['ratings_disabled']=invideos_df['ratings_disabled'].replace({True:1,False:0})
invideos_df['video_error_or_removed']=invideos_df['video_error_or_removed'].replace({True:1,False:0})
```

```python
#checking some sample values after some changes
invideos_df.sample(10)
```

```python
#to check the datatype format has been changed
invideos_df.info()
```

## Exploratory Analysis and Visualization


Q.plot a linegraph to know how the number of total views in youtube has varied from November 2017 to june 2018

```python
#checking how many months are there in dataset (-1 means november and 0 means december)
invideos_df.month.unique()
```

```python
#calculating views in each month
monthly_videos = invideos_df.groupby('month')['views'].sum()
monthly_videos
```

```python
#ploting every month views as line graph 
plt.plot(monthly_videos);
plt.xlabel('Months')
plt.ylabel('Total views(in billion)');
```

Q.plot a histogram to see in which month there was highest number of trending video uploads

```python
#calculating the no. of uploads in every month
more_uploads_df = invideos_df.month.value_counts()
more_uploads_df
```

```python
#ploting the histogram on no. of uploads
plt.title("Distribution of trending videos in the different months")
plt.hist(invideos_df.month);
```

In the month of december 2017 there was highest number of video uploads.

Q.Plot the bargraph of the most 10 disliked videos

```python
#selecting the top 10 disliked videos
most_disliked_df = invideos_df.sort_values('dislikes',ascending=False).head(10)
```

```python
#top 10 disliked videos are
most_disliked_df
```

```python
#ploting the barplot for the disliked videos
plt.figure(figsize=(12, 12))
sns.barplot(most_disliked_df.dislikes, most_disliked_df.title)
plt.ylabel("Title of the Videos");
plt.xlabel('No. of dislikes');
plt.title('Most disliked Videos');
```

Q.Does the videos which has highest number of likes will have lowest number of dislikes

```python
#Scatterplot to show how the dislikes and likes are distributed
plt.title('correlation of likes and dislikes for videos') 
sns.scatterplot(invideos_df.likes, invideos_df.dislikes, s=100);
```

Not compulsorily,there is one video which has highest number of likes and dislikes in the above scatterplot.
but maximum of videos which has highest number of likes, will have less number of dislikes

At what time of the day there was more upload of videos, plot a bargraph on that

```python
#calculating how many videos are uploaded at each hour
invideos_hour = invideos_df.hour.value_counts()
invideos_hour
```

```python
#bar plot to show Distribution of videos published at different hours
plt.figure(figsize=(12,6))
plt.title("Distribution of videos published at different hours")
sns.barplot(invideos_hour.index, invideos_hour);
```

At 2pm of the day there was highest number of uploads.

# Asking and Answering Questions

Q.At which time the video which has max likes got published

```python
#selecting the highest likes from data frame
highest_liked = invideos_df.likes.max()
highest_liked
```

```python
#selecting the publish time of the highest liked video
highest_liked_publish_time = invideos_df[invideos_df.likes == highest_liked].publish_time
highest_liked_publish_time
```

At 1.26pm the video with max likes got published

Q.In how many videos the video errors are removed..?

```python
#counting in how many videos there are no errors
video_error_removed=invideos_df.video_error_or_removed.count()
video_error_removed
```

Almost,In all videos the errors got removed.

Q.serch how many videos are tagged to Music

```python
#selecting all the tags which has keyword music
music_count=invideos_df['tags'].str.contains('music').sum()
if music_count>0:
    print ("There are {m} musical Videos".format(m=music_count))
```

Around 588 videos are related to music.
